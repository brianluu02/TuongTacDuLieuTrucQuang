```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("D:\\Nam_3_ki_2\\tuong tac du lieu\\w12\\data.csv")
```

```python
X = df['Age'].values.reshape(-1, 1)
y = df['BoughtBurger'].values
```

```python
model = LogisticRegression()
model.fit(X, y)
```

```python
x_values = np.linspace(0, 80, 100)
y_values = model.predict_proba(x_values.reshape(-1, 1))[:, 1]
```

```python

plt.scatter(df['Age'], df['BoughtBurger'], color='blue')
plt.plot(x_values, y_values, color='red')
plt.xlabel('Age')
plt.ylabel('Bought Burger?')
plt.show()
```
