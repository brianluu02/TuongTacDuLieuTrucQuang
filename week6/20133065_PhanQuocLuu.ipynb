```python
def get_beta(X, Y):
    # Tính toán hệ số beta
    beta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)
    return beta
```

```python
X_ones = X * np.ones((200, 2))
X_ones[:, 0] = 1
# print(X_ones)
coef = get_beta(X_ones, Y)
print(coef)
```
