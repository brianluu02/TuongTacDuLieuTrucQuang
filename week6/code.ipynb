### Bài 4

```python
import plotly.express as px
import numpy as np

# đặt giá trị mục tiêu là số tùy ý 
y = 50

# Tạo ra 1000 giá trị ngẫu nhiên từ phân phối đều theo công thức đã cho của Z 
z = np.random.uniform(y-5, y+5, 1000)

# Tạo biểu đồ histogram từ các giá trị Z vừa tính ngẫu nhiên
# trục x sẽ hiển thị các giá trị z và trục y là tần suất xuất hiện của z
fig = px.histogram(x=z, nbins=30, opacity=0.9,
                   labels={'x': 'truc_x', 'y': 'truc_y'})

# thêm đường thẳng dọc trên biểu đồ histogram tại vị trí 𝑦 ban dau
fig.add_vline(x=y, line_dash="dash", line_color="yellow",
              annotation_text="Giá trị mục tiêu của y"
              )

# chú thích khi di chuột vào các cột trong biểu đồ histogram
fig.update_traces(hovertemplate="giá trị của Z là : %{x}")

# Display the plot
fig.show()
```
