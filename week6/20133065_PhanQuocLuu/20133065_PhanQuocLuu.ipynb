# Bài 4


```python
import numpy as np
import plotly.graph_objs as go
import plotly.offline as pyo

y = np.linspace(0, 300, 100)
z = np.random.uniform(low=y-5, high=y+5, size=100)

hover_text = []
for i in range(len(y)):
    hover_text.append('y:{:.2f}, z:{:.2f}'.format(y[i], z[i]))

trace = go.Scatter(x=y, y=z, mode='markers', text=hover_text, hoverinfo='text')

layout = go.Layout(title='Biểu đồ dạng động Plottly',
                   xaxis=dict(title='y'),
                   yaxis=dict(title='z'),
                   hovermode='closest')

fig = go.Figure(data=[trace], layout=layout)

mean_y = np.mean(y)
mean_z = np.mean(z)
a = 1
b = mean_z - a * mean_y
z_predict = a * y + b

target_trace = go.Scatter(x=y, y=z_predict, mode='lines', line=dict(color='red', width=2))
fig.add_trace(target_trace)

fig.show()
```

```python
def entropy(x):
    hist, _ = np.histogram(x, bins=10)
    p = hist / np.sum(hist)
    return -np.sum(p * np.log2(p + 1e-12))

target_trace = np.random.normal(size=1000)
h = entropy(target_trace)
print(h)
```
